## 4\. Gradebook: Aggregations and broadcasting

Reuse the “Gradebook” idea from class and extend it.

```py
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])
```

1. Compute each student’s **average score** (per row).  
2. Compute each subject’s **average score** (per column).  
3. Create `scores_centered` by subtracting the **subject mean** from each column (broadcasting).  
4. For `scores_centered`, compute per-student averages again.  
5. Compare which student looks best by **raw average** vs **centered average**.  
6. In a markdown cell, explain briefly why centered scores might give a fairer comparison.

Q5 Answer:
Student 3 looks best for both cases.

Q6 Answer:
Centering scores removes the effect of subject difficulty by measuring each score relative to the class average for that subject, rather than its absolute value. A high score in an easy subject contributes less (because everyone scores high), while a high score in a hard subject contributes more (because few others can match it). This makes the comparison fairer — a student can no longer inflate their average simply by being strong in easier subjects.
---

````python
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])

#1
average_score_row = np.average(scores, axis=1)
print(f"average score per row: \n {average_score_row} \n")

#2
average_score_column = np.average(scores, axis=0)
print(f"average score per column: \n {average_score_column} \n")

#3
scores_centered = scores - average_score_column
print(f"scores centered: \n {scores_centered} \n")

#4
per_student_avg = np.average(scores_centered, axis=1)
print(f"score centered per-student average: \n {per_student_avg}")

````

```py

    155, 145, 135, 125, 115
])
```

1. Reshape `daily_visits` into a `(6, 5)` array representing **6 weeks × 5 days**.  
2. Compute **total visits per week**.  
3. Compute **average visits per day of the week** (e.g., all “day 1 of week” together, all “day 2 of week” together, etc.).

````python
daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])

#1
daily_visit_by_week = daily_visits.reshape(6, 5)
print(f"daily visit by week: \n {daily_visit_by_week} \n")

#2
total_visits_per_week = daily_visit_by_week.sum(axis = 1)
print(f"total visits per week: \n {total_visits_per_week} \n")

#3
average_visits_per_day_of_the_week = daily_visit_by_week.sum(axis = 0)
print(f"average visits per day of the week: \n {average_visits_per_day_of_the_week} \n")

#4
original_array = daily_visit_by_week.flatten()
print(f"flatten back to 1D: \n {original_array} \n")
print(f"Does it match the original array? {np.array_equal(daily_visits, original_array)}")


````

## 6\. Mini-project: Simple scoring model

This mirrors the tiny matrix multiply example from class, but with more features.

```py
# Each row: [page_views, time_on_site (minutes), past_purchases]
X = np.array([
    [10,  3.5,  0],
    [25,  5.0,  1],
    [40,  2.0,  0],
    [15, 10.0,  3],
    [30,  4.0,  2]
])

customers = np.array(["C1", "C2", "C3", "C4", "C5"])
```

1. Choose a weight vector `w = [w_views, w_time, w_purchases]` (for example `[0.1, 0.5, 1.0]`).  
2. Compute a **score** for each customer using `scores = X @ w`.  
3. Rank customers by score (highest first).  
4. Change the weights to emphasize **past\_purchases** more than other features, and recompute.  
5. In a markdown cell, answer:  
   - Which customer is top-ranked before vs after changing weights?  
   - In what real-world situation might you prefer each weighting?

Q5 Answer:

The ranking depends entirely on the weights chosen. With purchase-heavy weights, C4 ranks first due to its high past_purchases. With balanced weight slightly emphasizing page_views and time_on_site (e.g. [0.5, 1.0, 0.1]), C3 is pushed to the top. This shows that there is no universally "correct" weighting — the right choice depends on the business objective. If the goal is acquiring new engaged users, weight page_views and time_on_site higher. If the goal is retaining proven buyers, weight past_purchases higher.
---

````python
# Each row: [page_views, time_on_site (minutes), past_purchases]
X = np.array([
    [10,  3.5,  0],
    [25,  5.0,  1],
    [40,  2.0,  0],
    [15, 10.0,  3],
    [30,  4.0,  2]
])

customers = np.array(["C1", "C2", "C3", "C4", "C5"])

#1 balanced wright with time_on_site slightly higher
w_1 = np.array([0.5, 1.0, 0.1])

#2
scores_1 = X @ w_1
print(scores_1)

#3
ranked_indices_1 = np.argsort(-scores_1)
ranked_customers_1 = customers[ranked_indices_1]
print(ranked_customers_1)
print()

#4
w_2 = np.array([0.1, 0.05, 5.0])
scores_2 = X @ w_2
print(scores_2)
ranked_indices_2 = np.argsort(-scores_2)
ranked_customers_2 = customers[ranked_indices_2]
print(ranked_customers_2)

````

````python
X = np.array([
    [10,  3.5,  0],
    [25,  5.0,  1],
    [40,  2.0,  0],
    [15, 10.0,  3],
    [30,  4.0,  2]
])
customers = np.array(["C1", "C2", "C3", "C4", "C5"])

# Weight set 1: Balanced
w1 = np.array([0.1, 0.05, 1.0])
scores1 = X @ w1

# Weight set 2: Heavy on Purchases
w2 = np.array([0.05, 0.1, 5.0])
scores2 = X @ w2

print("Balanced Rankings:", customers[np.argsort(-scores1)])
print("Purchase-heavy Rankings:", customers[np.argsort(-scores2)])
````
